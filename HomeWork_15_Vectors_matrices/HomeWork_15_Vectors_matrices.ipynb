Изобразите с помощью matplotlib. Изобразите точку x + 2y + 3z. Найдите угол между векторами x, y и x, z.

```python
x = np.array([1,1])
y = np.array([2, 0])
z = np.array([0,2])
```

```python
###Найдите угол между векторами x, y и x, z.
```

```python
def cosine( x, y ):
    """
    Подсчет косинуса угла между векторами x, y по их координатам
    """
    
    # длины векторов
    aLength = np.linalg.norm( x )
    bLength = np.linalg.norm( y )
    
    return np.dot( x, y ) / ( aLength * bLength )

#cosine( x, y )

np.arccos( cosine( x, y ) ) * 360 / 2 / np.pi # угол между векторами в градусах
```

```python
def cosine( x, z ):
    """
    Подсчет косинуса угла между векторами x, z по их координатам
    """
    
    # длины векторов
    aLength = np.linalg.norm( x )
    bLength = np.linalg.norm( z )
    
    return np.dot( x, z ) / ( aLength * bLength )

#cosine( x, z )

np.arccos( cosine( x, z ) ) * 360 / 2 / np.pi # угол между векторами в градусах
```

##Задание 2
Найдите собственные значения и собственные вектора матриц (необходимо решение на numpy и решение по алгоритму на бумажке). Для матрицы 3x3 можно посмотреть на корни характеристического многочлена, посчитанные в numpy.

```python
m1 = np.array([[2, 2],
               [1, 3]])

m2 = np.array([[4, 1, -1],
               [1, 4, -1],
               [-1, -1, 4]])
```

```python
w, v = np.linalg.eig(m1)

for i in range(len(w)):
    print("Собственное значение " + str(w[i]))
    print("соответствующий ему собственный вектор " + str(v[:,i]))
```

```python
w, v = np.linalg.eig(m2)

for i in range(len(w)):
    print("Собственное значение " + str(w[i]))
    print("соответствующий ему собственный вектор " + str(v[:,i]))
```
